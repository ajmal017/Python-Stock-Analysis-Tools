### Specifying the details about the given NSE based understanding to common terms


```python
def entries_transfer(keys,r):
    val={}
    for key in keys:
        if key in r.keys():
            val[key]=r[key]
    val['02.Last_Value_Google'] = val.pop('lastPrice')
    val['03.Price_change']= val.pop('change')
    val['04.Percentage_change']= val.pop('pChange')
    val['01.COMPANY']= val.pop('companyName')
    val['05.ClosePrice-End']= val.pop('closePrice')
    val['07.Max-Value-day']= val.pop('dayHigh')
    val['08.Low-Value-day']= val.pop('dayLow')
    val['06.Open-time-Price']= val.pop('open')
    val['10.52-Week-High']= val.pop('high52')
    val['11.52-Week-Low']= val.pop('low52')
    val['09.Average_Value-Stock']= val.pop('averagePrice')
    return val
```

### The following is the function to remove the values from the total list of values

```python
 def entries_to_remove(entries, r):
    entries=('indexVar','buyQuantity3','buyQuantity4','buyQuantity5','sellPrice2','sellPrice3') # Import any of the following values you want to remove
    for key in entries:
        if key in r.keys():
            del r[key]
        pprint(r)
    
```

```python
from pprint import pprint# just for neatness of display
a=['ceatltd','reliance','LT','sparc','SBIN','MRF','ONGC','rcom']
keys=('companyName','closePrice','lastPrice','dayHigh','dayLow','high52','low52','averagePrice','pChange','open','change','previousClose')
for x in a:
    q = nse.get_quote(x) # it's ok to use both upper or lower case for codes.
    r=dict(q)
    val=entries_transfer(keys,r);
     
    pprint(val)
```
